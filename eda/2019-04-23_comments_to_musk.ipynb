```python
for i, row in elon.iterrows():
    start_datetime = row.Datetime
    end_datetime = row.Datetime + pd.DateOffset(hours=0.5)
    
    tweet_id = row['id']
    
    start_month, start_year = "0" + str(start_datetime.month), start_datetime.year
    end_month, end_year = "0" + str(end_datetime.month), end_datetime.year
    
    file1 = "../../data/Comment_to_Elon/to_Elon_" + str(start_year) + "-" + start_month[-2:] + ".csv"
    file2 = "../../data/Comment_to_Elon/to_Elon_" + str(end_year) + "-" + end_month[-2:] + ".csv"
    
    if file1 == file2:
        
        comments = pd.read_csv(file1, sep=";", error_bad_lines=False) 
    
    else:
        comments1 = pd.read_csv(file1, sep=";", error_bad_lines=False)
        comments2 = pd.read_csv(file2, sep=";", error_bad_lines=False)
        comments = comments1.append(comments2)
        
    subset_df = comments[comments['origin']==np.float64(tweet_id)]
    
    if subset_df.shape[0] != 0:
        print(tweet_id)
        
        break
```
