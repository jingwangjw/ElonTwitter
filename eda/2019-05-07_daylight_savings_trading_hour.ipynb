```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
file = '../../data/features/2019_04_30_TSLA_y_02_yishuang.pickle'

tsla = pickle.load(open(file, "rb"))
```

```python
tsla.head()
```

```python
tsla["Start_Datetime"] = pd.to_datetime(tsla["Start_Datetime"])
tsla["End_Datetime"] = pd.to_datetime(tsla["End_Datetime"])
```

```python
daylight_dates = tsla[tsla.Trading_Hour==8].Start_Datetime
```

```python
daylight_dates = list(daylight_dates.dt.date)
```

```python
daylight_dates
```

```python
tsla['Date'] = tsla['Start_Datetime'].dt.date
tsla.head()
```

```python
tsla[['Start_Datetime', 'End_Datetime', 'Trading_Hour']].groupby('Trading_Hour').agg(['count'])
```

```python
tsla[tsla['Date'].isin(daylight_dates)]
```

```python
tsla[tsla['Date'].isin(daylight_dates)].count()/7
```

```python
new_trading_hours = tsla[tsla['Date'].isin(daylight_dates)]['Trading_Hour'] - 1
new_trading_hours
```

```python
tsla.loc[tsla['Date'].isin(daylight_dates), 'Trading_Hour'] = new_trading_hours
```

```python
tsla[['Start_Datetime', 'End_Datetime', 'Trading_Hour']].groupby('Trading_Hour').agg(['count'])
```

```python
tsla.head()
```

```python
tsla.columns
```

```python
tsla = tsla.drop(columns=['Date'])
```

```python
tsla.columns
```

```python
output_file = '../../data/features/2019_05_07_TSLA_y_03_yishuang.pickle'

pickle.dump(tsla, open(output_file, "wb"))
```
