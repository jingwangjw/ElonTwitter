```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import datetime
```

```python
#reading in elon's tweets 
df_elon=pd.read_csv("../../data/twitter/Elon_0402.csv")
```

```python
df_elon.tail(20)
```

```python
df_elon.info()
```

```python
#obtaining Datetime, year, month, day, and date columns of tweets
df_elon['DateTime'] = pd.to_datetime(df_elon['date'])
df_elon['year']= df_elon['DateTime'].dt.year
df_elon['month']=df_elon['DateTime'].dt.month
df_elon['day']=df_elon['DateTime'].dt.day
df_elon['date']=df_elon['DateTime'].dt.date
df_elon.head()
```

```python
#dropping NaN columns and selecting useful columns
df_elon_short=df_elon[["retweets","favorites","text","DateTime","year","month","day"]].dropna()
df_elon_short.head()
```

```python
#sort by retweets
df_elon_short_sort=df_elon_short.sort_values(by=["retweets"],ascending=False).reset_index(drop=True)
df_elon_short_sort.head()
```

# Investigate whether taking Tesla Private has impact on Tsla price

```python
#sort by retweets 
df_elon_short_sort['DATE'] = pd.to_datetime(df_elon_short_sort['DateTime'].dt.date)
df_private=df_elon_short_sort.sort_values(by=["retweets"],ascending=False)
df_private.head()
```

```python
#tweet of taking tesla private was from 2018-08-07
df_private[df_private["DATE"]=="2018-08-07"]
```

```python
#read in tsla stock price
df_tsla=pd.read_csv("../../data/stock_price/TSLA.csv")
```

```python
#obtaining datetime, year, month and day columns 
df_tsla['DateTime'] = pd.to_datetime(df_tsla['Date'])
df_tsla['year']= df_tsla['DateTime'].dt.year
df_tsla['month']=df_tsla['DateTime'].dt.month
df_tsla['day']=df_tsla['DateTime'].dt.day
```

```python
#want to see effect that taking tesla private would have on tsla price within two weeks after the tweet
df_tsla_private=df_tsla[(df_tsla['DateTime']<="2018-08-22") & (df_tsla['DateTime'] >="2018-08-06")]
df_tsla_private
```

```python
x=df_tsla_private["DateTime"]
y=df_tsla_private["Close"]

fig, ax = plt.subplots(figsize=(20,10))
ax.plot(x, y)
ax.set_title("TSLA Price Near Period When Elon Considered Taking Tesla Private", fontsize=30)
ax.set_xlim(('2018-08-06', '2018-08-21'))

ax.axvline(x='2018-08-07', color='red')


plt.show()
```
