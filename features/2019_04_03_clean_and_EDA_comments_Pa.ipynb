```python
import h5py
import pandas as pd
path= '../../data/features/2019_04_30_Eloncomments1hour_fix2018-04_Pa.hdf5'
f = pd.read_hdf(path,key='table')
f.head()
```

```python
# fix month 4/2018 tweets


for i,x in f.iterrows():
    
    if len(allfea[allfea.id == x.id]) > 0:
        
#         if len(allfea[allfea.id == x.id]) > 0
    
        if allfea[allfea.id == x.id].iloc[0].countComment <= x.countComment:
#             print(allfea[allfea.id == x.id])
            print(allfea[allfea.id == x.id].index.values)
#             print(allfea[allfea.id == x.id].index.values)
            allfea.at[allfea[allfea.id == x.id].index.values[0],'CommentSentimental'] = x.sentimental
            allfea.at[allfea[allfea.id == x.id].index.values[0],'CommentSubjectivity'] = x.subjectivity
            allfea.at[allfea[allfea.id == x.id].index.values[0],'retweets'] = x.retweets
            allfea.at[allfea[allfea.id == x.id].index.values[0],'favorites'] = x.favorites
            allfea.at[allfea[allfea.id == x.id].index.values[0],'countComment'] = x.countComment
            allfea.at[allfea[allfea.id == x.id].index.values[0],'posCommentAmount'] = x.posCommentAmount
            allfea.at[allfea[allfea.id == x.id].index.values[0],'negCommentAmount'] = x.negCommentAmount
            allfea.at[allfea[allfea.id == x.id].index.values[0],'neuCommentAmount'] = x.neuCommentAmount
            allfea.at[allfea[allfea.id == x.id].index.values[0],'posCommentSum'] = x.posCommentSum
            allfea.at[allfea[allfea.id == x.id].index.values[0],'negCommentSum'] = x.negCommentSum
            allfea.at[allfea[allfea.id == x.id].index.values[0],'CommentSD'] = x.CommentSD
            allfea.at[allfea[allfea.id == x.id].index.values[0],'posCommentSD'] = x.posCommentSD
            allfea.at[allfea[allfea.id == x.id].index.values[0],'negCommentSD'] = x.negCommentSD
#             print(allfea)
            print(allfea[allfea.id == x.id].countComment, x.countComment)
allfea.head(10)
```

```python
# fillna and clean na

allfea['negCommentPercent'] = allfea.negCommentAmount/allfea.countComment
allfea['posCommentPercent'] = allfea.posCommentAmount/allfea.countComment
# allfea['negCommentAvg'] = allfea.negCommentSum/allfea.countComment
# allfea['posCommentAvg'] = allfea.posCommentSum/allfea.countComment
allfea.head()
allfea.columns
for x in allfea.columns:

    if allfea[x].dtypes == object: 
        allfea[x] = allfea[x].fillna('')
    elif allfea[x].dtypes == float: 
        allfea[x] = allfea[x].fillna(0.0)
    elif allfea[x].dtypes == int: 
        allfea[x] = allfea[x].fillna(0)
        
allfea
```

```python
# these 3 features are all related to the number of comments (countComment) so I dropped them

allfea = allfea.drop(['posCommentAmount', 'negCommentAmount', 'neuCommentAmount'],1)
```
