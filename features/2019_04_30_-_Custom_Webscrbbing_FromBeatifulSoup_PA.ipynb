```python
def getoriginID(http,i):
    import time
    
    if(i%1000 == 0): 
        print(i)
    if(i%100 == 0): 
        time.sleep(3)
    if(http.isreply == True):
#         print(http.permalink)
        import requests
        from socket import error as SocketError
#     headers = {
#     'user-agent': 'curl/7.54.0'}
        try:
            page = requests.get(http.permalink, timeout=5)
            if(page.status_code == 200):

                from bs4 import BeautifulSoup as bs
                from bs4 import SoupStrainer
                strainer = SoupStrainer('form', attrs={'id': 'form1'})
                soup = bs(page.content, 'lxml', parse_only=strainer)

                lst = soup.find_all(class_="js-stream-item stream-item stream-item")
                if len(lst) > 0: 
                    id = lst[0]['data-item-id']
                    return(id)
                else:
                    return('')
            else:
                return('timeout')
                print('timeout ', http.permalink)
            #     for tag2 in soup.find_all(class_="js-stream-item stream-item stream-item"):
    #             print(tag)
        except SocketError as e:
            print(e)
            return('')
    else:
#         print('')
        return('')
```

```python
for i,f in file.iterrows():
    print('----------------------',f[0])
    month = int(f[0].split('-')[1].split('.')[0])
    
    comment = pd.read_csv('../../data/Comment_to_Elon/'+f[0], sep=';', error_bad_lines=False)

    comment.date = pd.to_datetime(comment.date)
    comment=comment[[True if c.isreply else False for i,c in comment.iterrows()]]
    comment = comment[[i.month ==month for i in comment.date]]

    comment['origin'] = [ getoriginID(c,j) for j,c in comment.iterrows()]

    with open('../../data/ProcessedComment/'+f[0].split('.')[0]+'.pickle', 'wb') as handle:
        pickle.dump(comment, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
