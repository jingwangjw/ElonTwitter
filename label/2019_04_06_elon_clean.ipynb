```python
pattern = re.compile('";(.*?);(.*?);(.*?);(.*?)"')
pattern2 = re.compile('"(.*?)"')
for i in range(len(index)):
    print(i)
    content = elon.iloc[index[i]]['text']
    result = re.findall(pattern, content)[0]
    new_content = content.replace('"'+re.findall(pattern2, content)[0]+'"','')
    elon.set_value(index[i],'permalink', elon.iloc[index[i]].geo)
    elon.set_value(index[i],'geo',result[0])
    elon.set_value(index[i],'mentions',result[1])
    elon.set_value(index[i],'hashtags',result[2])
    elon.set_value(index[i],'id',result[3])
    elon.set_value(index[i],'text',new_content)
    
```

# Add UTC to datetime

```python
# convert to datetime
elon["date"] = pd.to_datetime(elon["date"], utc=True)

# add timezone
elon = elon.rename(columns={'date':'Datetime'})
```

```python
elon.head(250)
```

# Remove useless columns

```python
# remove columns
elon = elon.drop(columns=['username', 'mentions', 'hashtags'])
```

# Filter only Tesla tweets

```python
# subset only tweets with "Tesla"
elon['text'] = elon["text"].str.lower()
elon_tesla = elon[elon["text"].str.contains("tesla")]
```

```python
elon_tesla.to_csv("../../data/twitter/elon_main.csv", index = False)
```

```python
# aggregate the data by date
```

```python
date = pd.to_datetime(elon.date).dt.date.unique()
```

```python
year = pd.to_datetime(elon.date).dt.year.unique()
```

```python
year
```

```python
elon['year'] = pd.to_datetime(elon.date).dt.year
```

```python
elon.groupby('year')['year'].count()
```

```python
elon['adj_date'] = pd.to_datetime(elon.date).dt.date
```

```python
a = elon.groupby('adj_date')['text'].apply('\n'.join).reset_index()
```

```python
b = elon.groupby(by='adj_date').agg({'retweets':sum}).reset_index()
c = elon.groupby(by='adj_date').agg({'favorites':sum}).reset_index()
```

```python
elon.fillna("", inplace = True)
d = elon.groupby('adj_date')['mentions'].apply(' '.join).reset_index()
e = elon.groupby('adj_date')['hashtags'].apply(' '.join).reset_index()
f = elon.groupby('adj_date')['permalink'].apply(' '.join).reset_index()
```

```python
new = pd.merge(a, b, on = 'adj_date')
new = pd.merge(new, c, on = 'adj_date')
new = pd.merge(new, d, on = 'adj_date')
new = pd.merge(new, e, on = 'adj_date')
new_elon = pd.merge(new, f, on = 'adj_date')
new_elon
```

```python
new_elon.to_csv("../../data/twitter/elon_aggregate.csv", index = False)
```
