```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle

import warnings
warnings.filterwarnings('ignore')
```

```python
file1 = open("../../data/features/2019_04_21_TSLA_y_02_yishuang.pickle",'rb')
tsla = pickle.load(file1, encoding='utf-8')

file2 = open("../../data/features/2019_04_23_nlp_jing_datetime_jeff.pickle",'rb')
elon = pickle.load(file2, encoding='utf-8')

tsla["Start_Datetime"] = pd.to_datetime(tsla["Start_Datetime"])
tsla["End_Datetime"] = pd.to_datetime(tsla["End_Datetime"])
elon["Datetime"] = pd.to_datetime(elon["Datetime"])
```

```python
tsla.head()
```

```python
elon[elon['TradingHour']==1].head()
```

```python
volumes = []

for i, row in elon.iterrows():

    tweet_datetime = row['Datetime']

    # get subset of tsla using tweet datetime
    subset_df = tsla[tsla['End_Datetime']>tweet_datetime]

    # get Total_Daily_Volume of first row of subset df
    volume = subset_df.iloc[0,:]['Volume']

    # append to list
    volumes.append(volume)

elon['that_hour_volume'] = pd.Series(volumes)
```
