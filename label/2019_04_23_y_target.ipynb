# y_target 1: volume

```python
# y_target 1: volume
volumes = []

for i, row in elon.iterrows():

    tweet_datetime = row['Datetime']
    day = row['DayofWeek']
    
    # Firday: if the twitter posted at the last 30 minutes, use the volume of the first hour of next Monday
    if day == 4:
        if tweet_datetime.minute < 30:
            # get subset of tsla using tweet datetime
            subset_df = tsla[tsla['End_Datetime']>tweet_datetime]
            # get hourly volume of first row of subset df
            volume = subset_df.iloc[0]['Volume']
        else:
            # get subset of tsla using tweet datetime
            subset_df = tsla[tsla['End_Datetime']>tweet_datetime]
            # get hourly volume of second row of subset df
            volume = subset_df.iloc[1]['Volume']
    
    # for Saturday and Sunday, we used the volume of the first hour of next Monday   
    if day > 4:
        # get subset of tsla using tweet datetime
            subset_df = tsla[tsla['End_Datetime']>tweet_datetime]
            # get hourly volume of first row of subset df
            volume = subset_df.iloc[0]['Volume']
    
    # for weekday, for the first 30 minutes, we use the volume of the current hour. For 30-45, we use the average of 
    # the current hour and the next hour. For the last 15 minutes, we use the volume of next hour.
    else:
        if tweet_datetime.minute < 30:
            # get subset of tsla using tweet datetime
            subset_df = tsla[tsla['End_Datetime']>tweet_datetime]
            # get hourly volume of first row of subset df
            volume = subset_df.iloc[0]['Volume']

        elif tweet_datetime.minute >= 30 & tweet_datetime.minute <= 45 & day != 4:
             # get subset of tsla using tweet datetime
            subset_df = tsla[tsla['End_Datetime']>tweet_datetime]
            # get the average volume of the first row and second row of subset df
            volume = (subset_df.iloc[0]['Volume'] + subset_df.iloc[1]['Volume'])/2

        else:
            # get subset of tsla using tweet datetime
            subset_df = tsla[tsla['End_Datetime']>tweet_datetime]
            # get hourly volume of second row of subset df
            volume = subset_df.iloc[1]['Volume']

    # append to list
    volumes.append(volume)

elon['That_hour_volume'] = pd.Series(volumes)
```
