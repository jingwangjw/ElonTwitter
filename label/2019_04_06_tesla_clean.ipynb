```python
pattern = re.compile('";(.*?);(.*?);(.*?);(.*?)"')
pattern2 = re.compile('"(.*?)"')
content = tesla.iloc[2632]['text']
result = re.findall(pattern, content)[0]
new_content = content.replace('"'+re.findall(pattern2, content)[0]+'"','')
tesla.set_value(2632,'permalink', tesla.iloc[2632].geo)
tesla.set_value(2632,'geo',result[0])
tesla.set_value(2632,'mentions',result[1])
tesla.set_value(2632,'hashtags',result[2])
tesla.set_value(2632,'id',result[3])
tesla.set_value(2632,'text',new_content)
```

```python
tesla.drop(['geo','id'],axis = 1, inplace = True)
tesla.drop(tesla[tesla.text.isna()].index, inplace = True)
```

```python
# aggregate the data by date
```

```python
tesla['adj_date'] = pd.to_datetime(tesla.date).dt.date
```

```python
a = tesla.groupby('adj_date')['text'].apply('\n'.join).reset_index()
b = tesla.groupby(by='adj_date').agg({'retweets':sum}).reset_index()
c = tesla.groupby(by='adj_date').agg({'favorites':sum}).reset_index()
tesla.fillna("", inplace = True)
d = tesla.groupby('adj_date')['mentions'].apply(' '.join).reset_index()
e = tesla.groupby('adj_date')['hashtags'].apply(' '.join).reset_index()
f = tesla.groupby('adj_date')['permalink'].apply(' '.join).reset_index()
```

```python
new = pd.merge(a, b, on = 'adj_date')
new = pd.merge(new, c, on = 'adj_date')
new = pd.merge(new, d, on = 'adj_date')
new = pd.merge(new, e, on = 'adj_date')
new_tesla = pd.merge(new, f, on = 'adj_date')
new_tesla
```

```python
new_tesla.to_csv("../../data/twitter/tesla_aggregate.csv", index = False)
```
