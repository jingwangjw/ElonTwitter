```python
import pandas as pd
stock = pd.read_csv("../../data/stock_price/TSLA_30m_2y.csv")
```

```python
stock['Date'] = stock['Datetime'].str[:11]
```

```python
avg_volume = stock.groupby('Date')['Volume'].mean().rename('Daily_avg_vol').reset_index()
```

```python
avg_price = stock.groupby('Date')['Close'].mean().rename('Daily_avg_close').reset_index()
```

```python
stock['Avg_price'] = (stock['High'] + stock['Low'])/2
```

```python
stock['SMA_2'] = stock.Close.rolling(window=2).mean()
```

```python
stock['SMA_3'] = stock.Close.rolling(window=3).mean()
```

```python
stock['SMA_4'] = stock.Close.rolling(window=4).mean()
```

```python
stock['SMA_5'] = stock.Close.rolling(window=5).mean()
stock['SMA_6'] = stock.Close.rolling(window=6).mean()
stock['SMA_7'] = stock.Close.rolling(window=7).mean()
```
